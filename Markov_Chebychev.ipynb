```python
import numpy as np
import scipy.stats as ss
from scipy.stats import poisson
from scipy.stats import expon
from scipy.stats import norm
from scipy.stats import binom
```

```python
X = poisson(5)

print('Prob(X >= 10) =', 1 - X.cdf(9))
```

```python
print(np.exp(-6))
```

```python
Y = expon(.1)

print('Prob(Y >= 60) =', 1 - Y.cdf(60))
```

```python
B = binom(100, .5)

print('Prob(B <= 35 or B>= 65) =', 1-B.cdf(64) + B.cdf(35))
```

```python
N = norm(0, 1)

print('Prob(|X|>= 1sigma) =', N.cdf(-1)+(1-N.cdf(1)))

print('Prob(|X|>= 2sigma) =', N.cdf(-2)+(1-N.cdf(2)))

print('Prob(|X|>= 3sigma) =', N.cdf(-3)+(1-N.cdf(3)))
```
