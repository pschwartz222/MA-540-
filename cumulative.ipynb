```python
import numpy as np
from numpy.linalg import eig
from numpy import linalg as LA
from numpy import random
from numpy.random import normal
import matplotlib.pyplot as plt
from sympy import diff, symbols
from sympy import cos, sin
import seaborn as sns
import scipy.stats as ss
```

```python
def F(x):
  pmf_X = [[0,1/4], [1,1/2], [2,1/4]] #probability mass function
  value_F = 0
  for point in pmf_X:
    if point[0] <= x:
      value_F = value_F + point[1]
  return(value_F)
```

```python
x = np.linspace(-1, 3, 500)
y = []

for X in x:
  y.append(F(X))

plt.plot(x, y, label='Prob X <= x')
plt.legend()
plt.show()
```

Now let's approximate a normal distribution.

```python
norm = normal(size=2000) #creates a random variable with approximate mean 0 and standard deviation 1.
sns.histplot(norm, kde=True)

plt.show()
```

Now let's plot the cumulative distribution function.

```python
def cumulative(Var, x): #input an array of values and a number x
  n = len(Var)
  number_below = 0
  for X in Var:
    if X <= x:
      number_below = number_below + 1
  prob = number_below/n
  return(prob)
```

```python
x_normal = np.linspace(-3, 3, 500)
y_normal = []

for X in x_normal:
  y_normal.append(cumulative(norm, X))
```

```python
plt.plot(x_normal, y_normal, label= 'Prob X <= x')
plt.legend()
plt.show()
```

Compare this to the actual Normal random variable

```python
x = np.linspace(-4, 4, 1000)
y = ss.norm.pdf(x)

plt.plot(x, y, label= 'probability density of N(0,1)')
plt.legend()
plt.show()
```

```python
#define x and y values to use for CDF
x = np.linspace(-4, 4, 1000)
y = ss.norm.cdf(x)

#plot normal CDF
plt.plot(x, y, label= 'cdf of N(0,1)')
plt.legend()
plt.show()
```
