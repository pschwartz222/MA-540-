```python
import numpy as np
from numpy.linalg import eig
from numpy import linalg as LA
import matplotlib.pyplot as plt
import sympy
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from sympy import diff, symbols
import pylab
import math
from sympy import cos, sin, exp, Matrix
```

```python
mu = [0, 0]
cov = [[2, 1], [1, 1]]

x_1, y_1 = np.random.multivariate_normal(mu, cov, 500).T
```

```python
plt.figure()
plt.scatter(x_1, y_1, label = 'data')
plt.xlabel('x_1')
plt.ylabel('y_1')
plt.title('Joint Guassian random variables')
plt.legend()
plt.show()
```

```python
X_1 = -y_1
Y_1 = x_1

plt.figure()
plt.scatter(X_1, Y_1, label = 'data')
plt.xlabel('X_1')
plt.ylabel('Y_1')
plt.title('Trasformation of x_3 and y_3')
plt.legend()
plt.show()
```
