```python
import random

def unbiased_coin_toss():
    # Generate a random number from 0 to 1
    x = random.uniform(0, 1)

    # Probability that the number falls between 0 and 0.5 is 1/2
    if x > 0.5:
        # Heads for True
        return True
    else:
        # Tails for False
        return False

# Test the outcome of the function
for i in range(10):
    print(unbiased_coin_toss())
```

```python
def coin_flip_sim(N):
    results= [] #list to store results (True/False)

    # Toss the coin N times and store the results in a list
    for i in range(N):
      result = unbiased_coin_toss()
      results.append(result)

    # Find the total number of Heads
    n_heads = sum(results)

    # Calculate the probability of Heads in the experiment
    probability_heads = n_heads / N

    print(f"Number of Heads: {n_heads}")
    print(f"Probability of Heads in {N} coin tosses: {probability_heads:.4f}")

coin_flip_sim(10000)
```

```python
def flip_the_coin():
    fair_coin = [0,1]
    return random.choice(fair_coin)

for i in range(10):
    print(flip_the_coin())
```

```python
def MonteCarlo_coin(N):
    results = [] #list to store results (0/1)
    #toss the coin N times and store the results in a list

    for i in range(N):
      result = flip_the_coin()
      results.append(result)

    #find the number of Heads
    n_heads = sum(results)

    #probability of Heads
    prob_heads = n_heads / N

    print(f"Number of Heads: {n_heads}")
    print(f"Probability of Heads in {N} coin tosses: {prob_heads:.4f}")
    return n_heads

def Stand_var_sim(M,N):
    exp_value = N / 2
    devs = [] #array to store variance in
    for i in range(M):
      sim_value = MonteCarlo_coin(N)
      sim_var = abs(sim_value - exp_value) #simulated deviation
      devs.append(sim_var)
      print(f"Deviation: {sim_var}")
      print( )
    sum_var = sum(devs)
    avg_var = sum_var / M
    print(f"approximate deviation: {avg_var: .2f}")

Stand_var_sim(20, 10000)
```
