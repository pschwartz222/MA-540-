```python
import numpy as np
from numpy.linalg import eig
from numpy import linalg as LA
import matplotlib.pyplot as plt
import sympy
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from sympy import diff, symbols
import pylab
import math
from sympy import cos, sin, exp, Matrix
```

```python
mu = [0, 0]
cov = [[1, .5], [.5, 1]]

x_1, y_1 = np.random.multivariate_normal(mu, cov, 500).T
```

```python
plt.figure()
plt.scatter(x_1, y_1, label = 'data')
plt.xlabel('x_1')
plt.ylabel('y_1')
plt.title('Joint Guassian random variables')
plt.legend()
plt.show()
```

```python
cov = [[2, .5], [.5, 1]]

x_2, y_2 = np.random.multivariate_normal(mu, cov, 500).T

plt.figure()
plt.scatter(x_2, y_2, label = 'data')
plt.xlabel('x_2')
plt.ylabel('y_2')
plt.title('Joint Guassian random variables')
plt.legend()
plt.show()
```

```python
cov = [[1, 0], [0, 1]]

x_3, y_3 = np.random.multivariate_normal(mu, cov, 500).T

plt.figure()
plt.scatter(x_3, y_3, label = 'data')
plt.xlabel('x_3')
plt.ylabel('y_3')
plt.title('Joint Guassian random variables')
plt.legend()
plt.show()
```

```python
X_1 = x_3 + y_3
Y_1 = x_3 - y_3

plt.figure()
plt.scatter(X_1, Y_1, label = 'data')
plt.xlabel('X_1')
plt.ylabel('Y_1')
plt.title('Trasformation of x_3 and y_3')
plt.legend()
plt.show()
```

```python

X_2 = x_1 + y_1
Y_2 = x_1 - y_1

plt.figure()
plt.scatter(X_2, Y_2, label = 'data')
plt.xlabel('X_2')
plt.ylabel('Y_2')
plt.title('Transformation of x_1 and y_1')
plt.legend()
plt.show()
```
