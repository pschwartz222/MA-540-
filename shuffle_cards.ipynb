```python
import numpy as np
from numpy.linalg import eig
from numpy import linalg as LA
import math
import random
```

```python
deck_cards = list(range(52))
```

```python
random.shuffle(deck_cards)
```

```python
North = [deck_cards[4*i] for i in range(13)]
South = [deck_cards[4*i+1] for i in range(13)]
East = [deck_cards[4*i+2] for i in range(13)]
West = [deck_cards[4*i+3] for i in range(13)]
```

```python
print(North)
print(South)
print(East)
print(West)
```

```python
def count_spades(hand):
  num_spades = 0
  for i in range(13):
    if int(hand[i]/13) < 1:
      num_spades = num_spades + 1
  return num_spades
```

```python
num_spades_north = count_spades(North)
num_spades_south = count_spades(South)
comb_spades = num_spades_north + num_spades_south
print(comb_spades)
```
