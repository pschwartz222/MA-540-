```python
import numpy as np
from numpy.linalg import eig
from numpy import linalg as LA
import math
import matplotlib.pyplot as plt
import scipy.stats as ss
from scipy.stats import norm
from scipy.stats import binom
```

Calculate the pmf of a binomial random variable

```python
def Binom_pmf(n,p,k): #calculates the probability mass function of a binomial random variable
  q = 1-p
  x = math.comb(n,k)*(p**k)*(q**(n-k))
  return(x)
```

```python
sum = 0
for i in range(11):
  sum = sum + Binom_pmf(10,1/2, i)
print(sum)  #prints 1.0 since the sum of all probabilities is 1
```

```python
print(f"{Binom_pmf(58,1/2,27): .4f}")
```

Let's graph the pmf.

```python

X = np.linspace(0, 10, 11) #row matrix with all integers 0 through 10
Y=[]
for k in range(11):
  Y.append(Binom_pmf(10,1/2,k))

plt.scatter(X,Y, label='pmf of B(10,1/2)')
plt.xlabel('k')
plt.ylabel('Prob(X=k)')
plt.legend()
plt.show()
```

Now let's calculate the cdf of the binomial random variable

```python
def Binom_cdf(n,p,x): #calculates the cumulative distribution function of a binomial random variable
  q = 1-p
  k = 0
  sum = 0
  while k <= x:
    sum = sum + math.comb(n,k)*(p**k)*(q**(n-k))
    k = k + 1
  return(sum)
```

```python
print(Binom_cdf(10,1/2,10))
```

Let's graph the cdf

```python
x = np.linspace(-1, 12, 500)
y = []

for t in x:
  y.append(Binom_cdf(10,1/2,t))

plt.plot(x, y, label='CDF of B(10,1/2)')
plt.xlabel('x')
plt.ylabel('Prob X<= x')
plt.legend()
plt.show()
```

```python
#define x and y values to use for CDF
x = np.linspace(-4, 4, 1000)
y = ss.norm.cdf(x)

#plot normal CDF
plt.plot(x, y, label= 'cdf of N(0,1)')
plt.legend()
plt.show()
```

```python
rv = binom(58,1/2)

print(f"{rv.pmf(27): .4f}") #probability that X = 27

prob = rv.cdf(32)-rv.cdf(25) #probability random variable is between 26 and 32.

print(f"{prob: .4f}")
```
